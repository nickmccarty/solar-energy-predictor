```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/processed/selected_features.csv', index_col = 0)
```

```python
df['Billing Period Duration'] = pd.Series([int(row.split(' ')[0]) for row in df['Billing Period Duration']])
```

```python
df = df.rename({'Billing Period Duration' :
                'Billing Period Duration (Days)'},
                axis = 1)
```

```python
columns_to_drop = ['Net Usage',
                   'Net Generation',
                   'Total Adjustment per kWh',
                   'Usage Charge',
                   'Billing Period Start']

df = df.drop(columns_to_drop, axis = 1)

df.set_index('Billing Period End', inplace = True)
```
