```python
# Importing dependencies

import pandas as pd
import glob
```

# Data Collection

<img src = 'https://raw.githubusercontent.com/nickmccarty/solar-energy-predictor/main/images/tesla-app-screenshot.jpg' width = 30%>

Thanks, Tesla, for making me download a year's worth of solar data one week at a time using your mobile app -- said no one ever...nevertheless, we persist and read them all into a single data structure for processing.

```python
# Looking in the data folder for the raw solar data CSVs,
# then concatenating them into a dataframe

df = pd.concat(map(pd.read_csv, glob.glob("../../data/raw/solar/*")))

# Casting `Date time` column as `datetime` object

df['Date time'] = pd.to_datetime(df['Date time'])

# Instantiating sorted df

df = df.sort_values(by='Date time')
```
